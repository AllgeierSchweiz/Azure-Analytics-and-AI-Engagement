```python
# Read from Cosmos DB analytical store into a Spark DataFrame and display 10 rows from the DataFrame
# To select a preferred list of regions in a multi-region Cosmos DB account, add .option("spark.cosmos.preferredRegions", "<Region1>,<Region2>")

df = spark.read\
    .format("cosmos.olap")\
    .option("spark.synapse.linkedService", "MediaCosmosDb")\
    .option("spark.cosmos.container", "videoindexerinsights")\
    .load()

display(df.limit(10))
```

```python
indexer = df.toPandas()
indexer.columns
```

```python
videos = indexer['videos'][0][0]
videos.__fields__
```

```python
len(indexer['videos'][1][0])
```

```python
row = videos['insights']
row.__fields__
```

```python
len(row['transcript'])   #[0]['text']
```

```python
# Extracting transcript text
number_of_videos = len(indexer['videos'])
extracted_text = []

tmp = ''
for i in range(number_of_videos):
    number_of_transcripts = len(indexer['videos'][i][0])
    for j in range(number_of_transcripts):
        try:  
            tmp = tmp + indexer['videos'][i][j]['insights']['transcript'][0]['text']
        except: 
            continue
    extracted_text.append(tmp)
    tmp = ''

extracted_text
```
